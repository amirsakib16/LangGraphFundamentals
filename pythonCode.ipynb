```python
class TemperatureConverter:
    def __init__(self):
        self.temperature = None
        self.state = None
    def temperature_converter(self, temp, from_unit, to_unit):
        if from_unit == "C" and to_unit == "F":
            result = (temp * 9/5) + 32
            print(f"{temp}°C is equal to {result}°F")
            self.temperature = result
            self.state = "F"
            return result
        elif from_unit == "F" and to_unit == "C":
            result = (temp - 32) * 5/9
            print(f"{temp}°F is equal to {result}°C")
            self.temperature = result
            self.state = "C"
            return result       
        else:
            raise ValueError("Invalid units. Use 'C' for Celsius and 'F' for Fahrenheit.")
    def comment(self):
        if self.state == "C" and  self.temperature >= 100:
            print("TOO MUCH HOT")
        elif self.state == "C" and self.temperature <= 100:
            print("VALID TEMPERATURE")
        elif self.state == "F" and self.temperature >= 212:
            print("TOO MUCH HOT")
        elif self.state == "F" and self.temperature <= 212:
            print("VALID TEMPERATURE")
TC = TemperatureConverter()
TC.temperature_converter(100, "C", "F")  
TC.comment()
TC.temperature_converter(212, "F", "C")  
TC.comment()
TC.temperature_converter(0, "C", "F")    
TC.comment()
TC.temperature_converter(32, "F", "C")  
TC.comment()
```
